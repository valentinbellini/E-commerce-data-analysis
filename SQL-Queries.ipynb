```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('e_commerce_db.db')

# TOP 10 Sales Products by Quantity Sold
query_top_products = """
SELECT 
    Description, 
    SUM(Quantity) AS TotalUnitSales
FROM 
    Sales
GROUP BY 
    Description
ORDER BY 
    TotalUnitSales DESC
LIMIT 10;
"""

# TOP 10 Sales Products by Total Sales Amount
query_top_productSalesAmount = """
SELECT 
    Description, 
    SUM(TotalPrice) AS TotalSalesAmount,
    SUM(Quantity) AS TotalUnitsSold,
    AVG(UnitPrice) AS AverageUnitPrice
FROM 
    Sales
GROUP BY 
    Description
ORDER BY 
    TotalSalesAmount DESC
LIMIT 10;
"""

# Execute the query and save the result in a new Pandas DataFrame
df_top_products_sql = pd.read_sql(query_top_products, conn)
df_top_productSalesAmount_sql = pd.read_sql(query_top_productSalesAmount, conn)

# Close the database connection
conn.close()

df_top_productSalesAmount_sql.head(10)
```
